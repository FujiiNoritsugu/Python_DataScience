### Numpyで100個のランダムな値のリストを作る

```python
import numpy as np
data = np.random.rand(100)
```

```python
data
```

### 作成したリストをPandasのSeriesに変換する。

```python
import pandas as pd
data_series = pd.Series(data)
```

```python
data_series
```

### PandasのSeriesをDataFrameに変換する。

```python
data_df = pd.DataFrame({'a':data_series})
```

```python
data_df
```

### DataFrameのデータの基本統計量を作成する。

```python
data_stat = data_df.describe()
```

```python
data_stat
```

### 基本統計量をCSVに出力する

```python
data_stat.to_csv('stat.csv')
```

### 100個のデータの線グラフを描く

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.array([i for i in range(100)])
plt.plot(x, data, '-')
plt.show()
```

### 作成したグラフを図として出力する

```python
x = np.array([i for i in range(100)])
fig = plt.figure()
plt.plot(x, data, '-')
plt.show()
fig.savefig('answer.png')
```
